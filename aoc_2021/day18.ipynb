```python
import math
    
class Node:
    def __init__(self, data=None, left=None, right=None):
        self.left = left
        self.right = right
        self.data = data
        
    def show(self):
        if self.data is not None:
            return str(self.data)
        else:
            return '['+self.left.show()+','+self.right.show()+']'
            
    def explode(self, d, pos):
        dp = 0
        out = []
        if self.data is not None:
            if d > 4:
                return 1, [pos]
            else:
                return 1, []
        ddp, do = self.left.explode(d+1, pos)
        dp += ddp
        pos += dp
        out.extend(do)
        ddp, do = self.right.explode(d+1, pos)
        dp += ddp
        out.extend(do)
        return dp, out
    
    def split(self):
        if self.data is not None:
            if int(self.data) > 9:
                self.left = Node(data=int(self.data)//2)
                self.right = Node(data=math.ceil(int(self.data)/2))
                self.data = None
            else:
                return False
            return True
        if not self.left.split():
            if not self.right.split():
                return False
            else: return True
        return True
    
    def mag(self):
        if self.data is not None:
            return int(self.data)
        return 3 * self.left.mag() + 2 * self.right.mag()
    
    
def parse(line):
    if line[0] == '[':
        i=1
        left, di = parse(line[i:])
        i += di + 1
        right, di = parse(line[i:])
        i += di + 1
        return Node(left=left, right=right), i
    else:
        stop = 0
        while line[stop].isdigit():
            stop += 1
        return Node(data=line[:stop]), stop

    

def explode(node):
    ex = node.explode(0, 0)[1]
    if ex == []: return node, False
    to_explode = min(ex)

    nums = []
    segments = []

    isnum = False
    acc = ''
    for c in node.show():
        if isnum:
            if c.isdigit():
                acc += c
            else:
                nums.append(len(segments))
                segments.append(acc)
                acc = c
                isnum = False
        else:
            if c.isdigit():
                segments.append(acc)
                acc = c
                isnum = True
            else:
                acc += c
    segments += acc

    el = segments[nums[to_explode]]
    er = segments[nums[to_explode+1]]
    if to_explode > 0:
        segments[nums[to_explode - 1]] = str(int(segments[nums[to_explode - 1]]) + int(el))
    if to_explode+1 < len(nums)-1:
        segments[nums[to_explode + 2]] = str(int(segments[nums[to_explode + 2]]) + int(er))
    segments[nums[to_explode]] = '0'
    segments[nums[to_explode]+1] = '0'
    segments[nums[to_explode]+2] = '0'
    node_str = ''
    for s in segments:
        node_str += s
    node_str = node_str.replace('[000]', '0')
    node_str
    return parse(node_str)[0], True    

def add(n1, n2):
    res = Node(left=n1, right=n2)
    changing = True
    while changing:
        res, exploding = explode(res)
        while exploding:
            res, exploding = explode(res)
        
        changing = res.split()
    return res
```

```python
lines = open('day18.txt').read().splitlines()
nodes = []
for line in lines:
    nodes.append(parse(line)[0])

while len(nodes) > 1:
    a = nodes.pop(0)
    b = nodes.pop(0)
    c = add(a,b)
    nodes.insert(0,c)
c.show()
c.mag()
```
